```python
import pandas as pd

# Sample DataFrame
data = {'Street Address': ['123 Main St', 'Apt 4B', 'Suite 10C']}
mdf = pd.DataFrame(data)

# Custom function to process each element in the 'Street Address' column
def process_address(address):
    # Split the address into words
    words = address.split()

    # Process each word: capitalize first letter if not a number, leave as is otherwise
    processed_words = [word.capitalize() if not word.isdigit() else word for word in words]

    # Join the processed words back into an address
    processed_address = ' '.join(processed_words)

    return processed_address

# Apply the custom function to the 'Street Address' column
mdf['Street Address'] = mdf['Street Address'].apply(process_address)

# Display the updated DataFrame
print(mdf)
```
