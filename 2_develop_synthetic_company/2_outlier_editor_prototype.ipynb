```python
import pandas as pd
import panel as pn

# Load the DataFrame from the CSV file
df = pd.read_csv('outliers.csv')

# Create a Panel app
def outliers_app():
    # Create a CheckboxGroup for accept, deny, and punt
    checkbox_group = pn.widgets.CheckBoxGroup(options=['Accept', 'Deny', 'Punt'], inline=True)

    # Define a function to filter the DataFrame based on the selected checkboxes
    def filter_data(accepted, denied, punted):
        filtered_df = df.copy()
        if 'Accept' in accepted:
            filtered_df = filtered_df[filtered_df['Decision'] == 'Accept']
        if 'Deny' in denied:
            filtered_df = filtered_df[filtered_df['Decision'] == 'Deny']
        if 'Punt' in punted:
            filtered_df = filtered_df[filtered_df['Decision'] == 'Punt']
        return filtered_df

    # Create a function to update the displayed DataFrame based on checkbox selection
    def update_data(event):
        accepted = 'Accept' if 'Accept' in checkbox_group.value else ''
        denied = 'Deny' if 'Deny' in checkbox_group.value else ''
        punted = 'Punt' if 'Punt' in checkbox_group.value else ''

        filtered_df = filter_data(accepted, denied, punted)

        # Update the Panel DataFrame widget
        data_widget.object = filtered_df

    # Register the update_data function to be called when the checkbox value changes
    #checkbox_group.on_change(update_data)

    # Display the initial DataFrame
    initial_df = filter_data('Accept', 'Deny', 'Punt')

    # Create a Panel DataFrame widget
    data_widget = pn.widgets.DataFrame(initial_df)

    # Combine widgets into a layout
    layout = pn.Column(
        pn.Row(pn.panel('## Outliers Panel App', align='center'), checkbox_group),
        data_widget,
    )

    return layout

# Run the Panel app
outliers_app().show()
```
